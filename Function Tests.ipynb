```r
library(tidyverse)
library(lubridate)
library(MASS)
```

```r
snapper=read.csv("data/Fish_Data_Snapper.csv")
snapper
```

```r
new_col= function (a, b){
    a %>%
    group_by(CaseAcc) %>%
    subset(Measurement == b)
}
```

```r
data1 = new_col(snapper, "Fork length")
data1
```

```r
new_col5= function (a, b, c, d){
    a %>%
    group_by(b) %>%
    subset(c == d)
}
```

```r
data2 = new_col5 (snapper, CaseAcc, Measurement, "Fork length")
```

```r
data3 = new_col5 ("snapper", "CaseAcc", "Measurement", "Fork length")
```

```r
data4 = new_col5 (snapper, 1, 7, "Fork length")
```
