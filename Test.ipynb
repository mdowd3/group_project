```r
library(tidyverse)
library(lubridate)
library(MASS)
```

```r
snapper=read.csv("data/Fish_Data_Snapper.csv")
```

```r
Fork_length_cm=snapper%>%group_by(CaseAcc)%>%subset(Measurement == "Fork length")
Total_length_cm = snapper%>%group_by(CaseAcc)%>%subset(Measurement == "Total length")
Weight_carcass_g =snapper%>%group_by(CaseAcc)%>%subset(Measurement == "Weight carcass")
Weight_liver_g =snapper%>%group_by(CaseAcc)%>%subset(Measurement == "Weight liver")

new_data= Fork_length_cm %>%
    dplyr::select(-c("Measurement", "MeasurementUnit")) 

new_data2= cbind(new_data, Total_length_cm$MeasureValue,
               Weight_carcass_g$MeasureValue,
               Weight_liver_g$MeasureValue)


    colnames(new_data2) = c("CaseAcc",
                       "Animal",
                       "DateCollected",
                       "Age",
                       "Sex",
                       "Waypoint",
                       "Fork_length_cm",
                       "Total_length_cm",
                       "Weight_carcass_g",
                       "Weight_liver_g")



new_data2
```

```r
new_data3 = new_data2 %>% mutate(
        HSI = ((Weight_liver_g/Weight_carcass_g)*100))
new_data3
```
